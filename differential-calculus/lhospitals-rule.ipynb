```python
import numpy as np  
import matplotlib.pyplot as plt
```

```python
# Define x approaches 0
xmin, xmax = -2, 2
x = np.linspace(xmin, xmax)
```

```python
# 1. y = (cosx + 2x -1) / 3x 
y = (np.cos(x) + (2*x) - 1) / (3 * x) 
plt.plot(x, y) 
plt.axhline(y=(2/3), xmin=xmin, xmax=xmax, ls='--')
plt.axvline(x=0, ls=':')
plt.title("Function y = (cosx + 2x - 1) / (3x)")
```

```python
# 2. y = (e^x + e^-x - 2) / (1 - cos(2x)) 
y = (np.exp(x) + np.exp(-x) - 2) / (1 - np.cos(2*x))
plt.plot(x, y) 
plt.axhline(y=(1/2), xmin=xmin, xmax=xmax, ls='--')
plt.axvline(x=0, ls=':')
plt.title("Function y = (e^x + e^-x - 2) / (1 - cos(2x)) ")
```
